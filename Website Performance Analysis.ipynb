```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data-export (1).csv")
```

```python
df.head()
```

# cleaning data and data validation for the above data set

```python
# Remove Unnamed row and set 0th row as Column heading.
df.columns = df.iloc[0]
df.head()
```

```python
# Removing 0th repeated row.
# Modify Column names.
df = df.drop(index = 0).reset_index(drop = True)
df.columns = ["channel group", "DateHour", "Users","Sessions", "Engaged Session", "Average engagement time per session", "Engaged sessions per user", 
              "Events per session", "Engagement rate", "Event count"]
```

```python
df.head()
```

```python
# chek datatype of each columns.
df.info()
```

```python
# Change DateHour column dtype object to Date Time Format.
df["DateHour"] = pd.to_datetime(df["DateHour"], format="%Y%m%d%H", errors='coerce')
```

```python
# validate dtype.
df.info()
```

```python
# check also in dataset
df.head()
```

```python
# Users column to last column values are in numbers so we have to change dtype to numberical Data.
# add new column to store hour data.
numeric_cols = df.columns.drop(["channel group", "DateHour"])
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
df["Hour"] = df["DateHour"].dt.hour
```

```python
# see dataset changes are done or not.
df.head()
```

```python
# validate datatype
df.info()
```

```python
# Here DateSet is Neet cleaned and ready.
# see statistics of dataset.
df.describe()
```

# Website Sessions and Users Over Time.

```python
# set style to grid.
sns.set(style="whitegrid")
```

```python
# Plot graph
plt.figure(figsize=(10,5))
df.groupby("DateHour")[["Sessions","Users"]].sum().plot(ax=plt.gca())
plt.title("Sessions and users over time")
plt.xlabel("DateHour")
plt.ylabel("count")
plt.show()
```

# Total users by channel

```python
# Plot Bar graph
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="channel group" , hue="channel group", y="Users", estimator=np.sum, palette="viridis")
plt.title("Total Users by Channel")
plt.xticks(rotation=45)
plt.show()
```

```python
df.head()
```

# Average Engagement time by channel

```python
# Plot bar graph
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="channel group", hue="channel group", y="Average engagement time per session", estimator= np.mean, palette="magma")
plt.title("Avg Engagement Time by Channel")
plt.xticks(rotation=45)
plt.show()
```

```python
df.head()
```

# Engagement Rate Distribution by channel

```python
# Plot Box Plot Graph
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="channel group",hue ="channel group", y="Engagement rate", palette="coolwarm")
plt.title("Engagement Rate Distribution by Channel")
plt.xticks(rotation=45)
plt.show()
```

# Engaged vs non engaged sessions

```python
df.head()
```

```python
session_df = df.groupby("channel group")[["Sessions", "Engaged Session"]].sum().reset_index()
session_df["Non-Engaged"] = session_df["Sessions"] - session_df["Engaged Session"]
session_df_melted = session_df.melt(id_vars="channel group", value_vars=["Engaged Session", "Non-Engaged"])

# Plot Bar graph
plt.figure(figsize=(8, 5))
sns.barplot(data=session_df_melted, x="channel group", y="value", hue="variable")
plt.title("Engaged vs Non-Engaged Sessions")
plt.xticks(rotation=45)
plt.show()
```

# Traffic by hour and channel

```python
# Plot Heatmap graph
heatmap_data = df.groupby(["Hour", "channel group"])["Sessions"].sum().unstack().fillna(0)

plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap="YlGnBu", linewidths=.5, annot=True, fmt='.0f')
plt.title("Traffic by Hour and Channel")
plt.xlabel("Channel Group")
plt.ylabel("Hour of Day")
plt.show()
```

# Engagement rate vs sessions over time

```python
df.head()
```

```python
# Plot Graph
df_plot = df.groupby("DateHour")[["Engagement rate", "Sessions"]].mean().reset_index()

plt.figure(figsize=(10, 5))
plt.plot(df_plot["DateHour"], df_plot["Engagement rate"], label="Engagement rate", color="green")
plt.plot(df_plot["DateHour"], df_plot["Sessions"], label="Sessions", color="blue")
plt.title("Engagement Rate vs Sessions Over Time")
plt.xlabel("DateHour")
plt.legend()
plt.grid(True)
plt.show()
```
